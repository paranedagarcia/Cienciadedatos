## Ejemplo b치sico
Aqu칤 tienes un ejemplo b치sico de c칩mo crear un gr치fico de dispersi칩n utilizando Seaborn:

````python
import seaborn as sns
import matplotlib.pyplot as plt
import requests
import pandas as pd
from io import StringIO
import urllib3

# Suppress insecure request warnings when using verify=False below (only for local/example use).
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# Cargar un conjunto de datos de ejemplo desde el repositorio de seaborn usando requests
# Usamos verify=False para evitar el error de validaci칩n SSL en entornos locales.
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
resp = requests.get(url, verify=False)
resp.raise_for_status()
tips = pd.read_csv(StringIO(resp.text))

# Crear un gr치fico de dispersi칩n
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.title("Gr치fico de dispersi칩n de propinas")
plt.show()
````

En este ejemplo, cargamos un conjunto de datos de ejemplo llamado "tips" y creamos un gr치fico de dispersi칩n que muestra la relaci칩n entre la cuenta total y la propina, diferenciando los puntos por d칤a de la semana.
