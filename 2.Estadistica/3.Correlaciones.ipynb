#### Coeficiente de correlaci贸n de Pearson
El coeficiente de correlaci贸n de Pearson es una medida estad铆stica que eval煤a la fuerza y la direcci贸n de la relaci贸n lineal entre dos variables cuantitativas. Su valor oscila entre -1 y 1, donde:
- 1 indica una correlaci贸n positiva perfecta: a medida que una variable aumenta, la otra tambi茅n lo hace.
- -1 indica una correlaci贸n negativa perfecta: a medida que una variable aumenta, la otra disminuye.
- 0 indica que no hay correlaci贸n lineal entre las variables.

La f贸rmula para calcular el coeficiente de correlaci贸n de Pearson $( r )$ es:

$$
r = \frac{n(\sum xy) - (\sum x)(\sum y)}{\sqrt{[n\sum x^2 - (\sum x)^2][n\sum y^2 - (\sum y)^2]}}
$$

Donde:
- $ n $ es el n煤mero de pares de datos.
- $ x $ e $ y $ son las variables que se est谩n correlacionando.

El coeficiente de correlaci贸n de Pearson es sensible a valores at铆picos, por lo que es importante visualizarlos mediante diagramas de dispersi贸n antes de realizar el an谩lisis. Adem谩s, solo captura relaciones lineales, por lo que no debe usarse para evaluar relaciones no lineales.

Las condiciones que se deben de cumplir para que el coeficiente de correlaci贸n de Pearson sea v谩lido son:

* La relaci贸n que se quiere estudiar es de tipo lineal (de lo contrario, el coeficiente de Pearson no la puede detectar).

* Las dos variables deben de ser num茅ricas.

* Normalidad: ambas variables se tienen que distribuir de forma normal. En la pr谩ctica, se suele considerar v谩lido aun cuando se alejan moderadamente de la normalidad.

* Homocedasticidad: la varianza de  debe ser constante a lo largo de la variable . Esto se puede contrastar si en un scatterplot los valores de  mantienen la misma dispersi贸n en las distintas zonas de la variable .

#### Coeficiente de correlaci贸n de Spearman
El coeficiente de correlaci贸n de Spearman es una medida no param茅trica que eval煤a la relaci贸n mon贸tona entre dos variables. A diferencia del coeficiente de Pearson, que asume que la relaci贸n es lineal, Spearman se basa en los rangos de los datos en lugar de los valores brutos. Esto lo hace m谩s robusto frente a valores at铆picos y adecuado para datos ordinales.

La f贸rmula para calcular el coeficiente de correlaci贸n de Spearman $( \rho )$ es:

$$
\rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)}
$$

Donde:
- $ d_i $ es la diferencia entre los rangos de cada par de datos.
- $ n $ es el n煤mero de pares de datos.

El coeficiente de Spearman tambi茅n oscila entre -1 y 1, con interpretaciones similares a las del coeficiente de Pearson. Un valor de 1 indica una correlaci贸n mon贸tona positiva perfecta, -1 indica una correlaci贸n mon贸tona negativa perfecta y 0 indica que no hay correlaci贸n mon贸tona.

#### Coeficiente de correlaci贸n de Kendall
El coeficiente de correlaci贸n de Kendall es una medida no param茅trica que eval煤a la concordancia entre dos variables ordinales. A diferencia de Pearson y Spearman, que se centran en las relaciones lineales y mon贸tonas, respectivamente, Kendall se basa en la comparaci贸n de pares de observaciones para determinar la direcci贸n y la fuerza de la relaci贸n.

La f贸rmula para calcular el coeficiente de correlaci贸n de Kendall $( \tau )$ es:

$$
\tau = \frac{(n_c - n_d)}{\frac{1}{2}n(n-1)}
$$

Donde:
- $ n_c $ es el n煤mero de pares concordantes.
- $ n_d $ es el n煤mero de pares discordantes.
- $ n $ es el n煤mero total de observaciones.

El coeficiente de Kendall tambi茅n oscila entre -1 y 1, con interpretaciones similares a las de Pearson y Spearman. 
* Un valor de 1 indica una concordancia perfecta, 
* -1 indica una discordancia perfecta y 
* 0 indica que no hay concordancia.


```python
### 驴Cu谩ndo debo usar cada coeficiente de correlaci贸n?

- **Pearson**: salo cuando ambas variables sean num茅ricas, la relaci贸n sea lineal y los datos cumplan (aproximadamente) con la normalidad y homocedasticidad. Es sensible a valores at铆picos.
- **Spearman**: salo cuando la relaci贸n entre variables sea mon贸tona pero no necesariamente lineal, o cuando los datos sean ordinales o tengan valores at铆picos.
- **Kendall**: salo con variables ordinales o cuando quieras evaluar la concordancia entre rangos. Es m谩s robusto en muestras peque帽as y menos sensible a empates que Spearman.

En resumen:  
- Pearson para relaciones lineales y datos normales.  
- Spearman para relaciones mon贸tonas o datos no normales/ordinales.  
- Kendall para evaluar concordancia en rangos, especialmente con muestras peque帽as o muchos empates.

**La correlaci贸n no implica causalidad**

No podr铆amos terminar este tutorial sin mencionar una advertencia importante: la correlaci贸n no implica causalidad.

La correlaci贸n solo cuantifica la fuerza y la direcci贸n de la relaci贸n entre dos variables. Puede haber una fuerte correlaci贸n entre dos variables, pero esto no permite concluir que una causa la otra. Cuando las correlaciones fuertes no son causales, las llamamos correlaciones espurias.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

try:
    df = pd.read_csv('../data/newborn_health_monitoring_with_risk.csv', encoding='ascii')
    
    # Convert 'date' column to datetime
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
    print('Data loaded successfully. DataFrame shape:', df.shape)
except Exception as e:
    print('Error loading data:', e)
    raise
```

```python
# Crear un nuevo DataFrame con la 煤ltima fecha ('date' m谩s alta) para cada baby_id
df_unique = df.sort_values('date').groupby('baby_id').tail(1).reset_index(drop=True)
# drop columna age_days
df_unique = df_unique.drop(columns=['age_days'])
df_unique.head()
```

```python
# Analizar la correlaci贸n entre 'birth_weight_kg' y 'birth_length_cm'
correlacion = df_unique['birth_weight_kg'].corr(df_unique['birth_length_cm'])
print(f"Coeficiente de correlaci贸n de Pearson entre 'birth_weight_kg' y 'birth_length_cm': {correlacion:.3f}")

# Visualizaci贸n con scatterplot
sns.scatterplot(x='birth_weight_kg', y='birth_length_cm', data=df_unique)
plt.title("Correlaci贸n entre peso y longitud al nacer")
plt.xlabel("Peso(kg)")
plt.ylabel("Longitud(cm)")
plt.grid(True)
plt.show()
# Gr谩fico de distribuci贸n para 'birth_weight_kg'
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df_unique['birth_weight_kg'], kde=True, color='skyblue')
plt.title("Distribuci贸n del peso al nacer (kg)")
plt.xlabel("Peso (kg)")
plt.ylabel("Frecuencia")

# Gr谩fico de distribuci贸n para 'length_cm'
plt.subplot(1, 2, 2)
sns.histplot(df_unique['birth_length_cm'], kde=True, color='salmon')
plt.title("Distribuci贸n de la longitud al nacer (cm)")
plt.xlabel("Longitud(cm)")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

```python
# Calcular la matriz de correlaci贸n de Pearson para las columnas num茅ricas de df_unique
cor_matrix = df_unique.corr(numeric_only=True)
print("Matriz de correlaci贸n de Pearson (num茅ricas):")
print(cor_matrix)

# Visualizaci贸n de la matriz de correlaci贸n con un heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cor_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Matriz de correlaci贸n de Pearson entre variables num茅ricas")
plt.show()
```
