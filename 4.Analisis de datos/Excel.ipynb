# Manejo de planillas (estilo Excel) en Python
===============================================================
En base a la popularidad de las hojas de cálculo tipo Excel para la gestión y análisis de datos, Python ofrece varias bibliotecas que facilitan la manipulación de estos archivos. A continuación, se presentan algunas de las bibliotecas más utilizadas y ejemplos básicos de cómo usarlas.

1. **Pandas**: Es la biblioteca más popular para el análisis de datos en Python. Permite leer y escribir archivos Excel de manera sencilla.

   ```python
   import pandas as pd

   # Leer un archivo Excel
   df = pd.read_excel('archivo.xlsx')

   # Escribir un DataFrame a un archivo Excel
   df.to_excel('archivo_salida.xlsx', index=False)
   ```

2. **OpenPyXL**: Esta biblioteca es útil para leer y escribir archivos Excel 2010 xlsx/xlsm/xltx/xltm.

   ```python
   from openpyxl import Workbook

   # Crear un nuevo libro de trabajo
   wb = Workbook()
   ws = wb.active

   # Escribir datos en la hoja
   ws['A1'] = 'Hola'
   ws['B1'] = 'Mundo'

   # Guardar el archivo
   wb.save('archivo.xlsx')
   ```

3. **xlrd y xlwt**: Estas bibliotecas se utilizan para leer y escribir archivos Excel antiguos (xls).

   ```python
   import xlrd
   import xlwt

   # Leer un archivo Excel
   libro = xlrd.open_workbook('archivo.xls')
   hoja = libro.sheet_by_index(0)
   print(hoja.cell_value(0, 0))

   # Escribir en un archivo Excel
   libro_nuevo = xlwt.Workbook()
   hoja_nueva = libro_nuevo.add_sheet('Hoja1')
   hoja_nueva.write(0, 0, 'Hola Mundo')
   libro_nuevo.save('archivo_nuevo.xls')
   ```

Estas bibliotecas proporcionan una amplia gama de funcionalidades para trabajar con archivos de Excel en Python, lo que facilita la manipulación y análisis de datos en este formato tan utilizado.
