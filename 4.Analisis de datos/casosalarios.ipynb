#### 13. ¿Cuál es el sueldo total de JOSEPH DRISCOLL incluyendo beneficios?

```python
# Buscar  JOSEPH DRISCOLL y mostrar TotalPayBenefits
mask = df['EmployeeName'] == 'JOSEPH DRISCOLL'
tpb = df.loc[mask, 'TotalPayBenefits']

if tpb.empty:
    print("No se encontró JOSEPH DRISCOLL")
else:
    # Mostrar todos los valores encontrados (si hay más de uno)
    print("TotalPayBenefits para JOSEPH DRISCOLL:")
    print(tpb.to_string(index=False))
    if tpb.nunique() == 1:
        print("\nValor único:", tpb.iloc[0])
    else:
        print(f"\nRegistros encontrados: {len(tpb)}; valores únicos: {tpb.unique()}")
```

#### 14. ¿Cuál es el nombre de la persona mejor pagada? (beneficios incluidos)

```python
# Obtener mayor TotalPayBenefits
max_tpb = df['TotalPayBenefits'].max()
best_paid = df.loc[df['TotalPayBenefits'] == max_tpb, ['EmployeeName', 'JobTitle', 'TotalPayBenefits']]

print(f"Monto máximo (TotalPayBenefits): {max_tpb}")
best_paid.reset_index(drop=True)
best_paid
```
